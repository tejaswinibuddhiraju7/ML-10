```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
print(iris.DESCR)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = Axes3D(fig, elev=-140, azim=100)

#Perform PCA with 3 components
pca = decomposition.PCA(n_components=3)
new_data = pca.fit_transform(iris.data)

# Create labels
for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(new_data[iris.target == label, 0].mean(),
              new_data[iris.target == label, 1].mean() + 1.5,
              new_data[iris.target == label, 2].mean(), name,
              bbox=dict(edgecolor='w', facecolor='w'))
    
ax.scatter(new_data[:, 0], new_data[:, 1], new_data[:, 2], c=iris.target)
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
plt.show()
```
